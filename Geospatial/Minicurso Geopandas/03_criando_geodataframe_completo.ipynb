Vamos importa a biblioteca do pandas para carregar nossos dados:

Use a função do pandas __read_excel__ para ler os arquivos baixados:

```python
# ler os dados com pandas
# df_roubos = pd.read_excel('dados/dados_roubo_celular_sp_2020.xls') (DELIMITADOR: tab)

df_roubos = pd.read_excel('dados/dados_roubo_celular_sp_2020.xlsx')
```

```python
# mostrar os dados com a função head
df_roubos.head(5)
```

```python
# mostrar toda as colunas para ver se temos informações de latitude e longitude
df_roubos.columns
```

Como podemos ver temos dados de __latitude__ e __longitude__ nulos. Sem essas informações não temos como colocar os dados no nosso mapa, então vamos remover todos os dados em que a __latitude__ ou a __longitude__ são nulos.

```python
# vamos usar o dropna com parâmetro subset para remover os dados
df_roubos = df_roubos.dropna(subset=['LATITUDE', 'LONGITUDE'])

df_roubos[['LATITUDE', 'LONGITUDE']]
```

# 2. Criar o GeoDataFrame (GeoPandas)

Como vimos anteriormente o GeoDataFrame (gdf) é basicamente um Dataframe do pandas com uma coluna "geometry" com objetos da biblioteca Shapely. Para construir o nosso gdf, vamos criar a nossa coluna geometry, do tipo _Point_ utilizando os nossos dados de latitude e longitude.

```python
# importar a classe point da biblioteca Shapely
from shapely.geometry import Point
```

```python
# criar a nossa coluna geometry utilizando apply e função lambda
df_roubos['geometry'] = df_roubos.apply(lambda row: Point(row.LONGITUDE, row.LATITUDE), axis=1)
```

Vamos ver se a nossa coluna geometry foi criada corretamente.

```python
# mostrar os dados com a função head
df_roubos.head(5)
```

Agora vamos criar o nosso GeoDataFrame.

```python
# antes vamos o tipo do nosso dataframe
type(df_roubos)
```

```python
# importar a biblioteca do geopandas
import geopandas as gpd
```

```python
# criar o GeoDataFrame
gdf_roubos = gpd.GeoDataFrame(df_roubos, geometry='geometry')
```

```python
# vamos ver o tipo do nosso dataframe após a transformação
type(gdf_roubos)
```

Vamos verificar o nosso dataframe novamente para ver se houve alguma mudança

```python
# mostrar os dados com a função head
gdf_roubos.head(5)
```

Como podemos ver a estrutura do nosso gdf é a mesma do pandas dataframe, e isso nós dá uma grande vantagem para manipular os dados, pois todas as funções herdadas do pandas também funcionarão no nosso geopandas.

E se agente usar a função __plot()__ no nosso gdf, qual resultado teremos?

```python
# mostrar os dados usando a função plot
gdf_roubos.plot()
```

Parece que já temos um esboço dos nosso dados. Vamos carregar o nosso gdf das fronteiras de municípios e mostrar os dois dados em um só mapa.

```python
# carregar os dados SP_Municipios_2019
gdf_municipios_sp = gpd.read_file('dados/SP_Municipios_2019.shp')
```

```python
# mostrar os dados carregados com a função plot
gdf_municipios_sp.plot()
```

Como podemos mostrar os dois em um mesmo mapa? Vamos primeiro fazer da forma RAIZ!! Usando o matplotlib!

```python
# importar a biblioteca matplotlib
import matplotlib.pyplot as plt
```

```python
# mostrar os dados com matplotlib
fig, ax = plt.subplots(figsize=(15,15))

gdf_roubos.plot(ax=ax)
gdf_municipios_sp.plot(ax=ax, facecolor='white', edgecolor='black')

plt.show()
```

# 3. Operação de interseção entre geometrias

Parece que os nosso dados estão muito concentrados na região sudeste de SP. Vamos mostrar só a geometria da capital para ver se a nossa visualização fica melhor.

```python
# pegar somente os gdf da capital são paulo
gdf_sp = gdf_municipios_sp[gdf_municipios_sp['NM_MUN'] == 'São Paulo']

gdf_sp
```

```python
fig, ax = plt.subplots(figsize=(15,15))

gdf_roubos.plot(ax=ax)
gdf_sp.plot(ax=ax, edgecolor='black', facecolor="None")

plt.show()
```

De acordo com os nosso dados, realmente existe uma grande concentração dos nossos dados na região da capital, região metropolitana e na costa.

E para filtrar somente os dados registrados dentro da capital? Podemos fazer isso facilmente utilizando as funções do nosso geodataframe.

```python
# vamos primeiro pegar o polygon de são paulo
polygon_sp = gdf_sp.iloc[0].geometry
```

```python
polygon_sp
```

```python
gdf_roubos_capital = gdf_roubos[gdf_roubos.intersects(polygon_sp)]
```

Aqui nós recebemos uma mensagem de __warning__, vamos analisar a mensagem.

Vamos mostrar novamente os dados em um mapa.

```python
fig, ax = plt.subplots(figsize=(10,10))

gdf_roubos_capital.plot(ax=ax)
gdf_sp.plot(ax=ax, edgecolor='black', facecolor="None")

plt.show()
```

Agora vamos salvar os nossos dados para usar posteriormente.

```python
# salvando os dados da geometria da capital
gdf_sp.to_file("dados/municipio_são_paulo.json", driver="GeoJSON")

# salvando os dados de roubo em sp
gdf_roubos_capital.to_file("dados/dados_roubo_celular_capital.json", driver="GeoJSON")
```

Já temos os dados de são paulo prontos para criar os nossos mapas interativos.

# 4. Operação de União entre geometrias

Como vimos no mapa anterior, nós temos dados de roubo da capital São Paulo e também de outros municpíos, principalmente na região metropolitana. E se eu quiser separar toda a região da __Grande São Paulo__?

```python
# criar uma lista com todos os municípios de grande sp

lista_grande_sp = ["São Paulo","Arujá","Barueri","Biritiba Mirim","Caieiras","Cajamar",
                   "Carapicuíba","Cotia","Diadema","Embu","Embu-Guaçu","Ferraz de Vasconcelos",
                   "Francisco Morato","Franco da Rocha","Guararema","Guarulhos","Itapecerica da Serra",
                   "Itapevi","Itaquaquecetuba","Jandira","Juquitiba","Mairiporã","Mauá","Mogi das Cruzes",
                   "Osasco","Pirapora do Bom Jesus","Poá","Ribeirão Pires","Rio Grande da Serra","Salesópolis",
                   "Santa Isabel","Santana do Parnaíba","Santo André","São Bernardo do Campo",
                   "São Caetano do Sul","São Lourenço da Serra Suzano","Suzano","Taboão da Serra",
                   "Vargem Grande Paulista"]
```

```python
# vamos ver a quantidade de municípios da nossa lista
len(lista_grande_sp)
```

```python
# vamos separar todos os municípios do nosso gdf que estão na nossa lista
gdf_grande_sp = gdf_municipios_sp[gdf_municipios_sp['NM_MUN'].isin(lista_grande_sp)]
```

```python
# vamos ver a quantidade de municípios que conseguimos separar
len(gdf_grande_sp)
```

```python
#
gdf_grande_sp = gdf_grande_sp.dissolve(by='SIGLA_UF')

# vamos mostrar no mapa os municípios que conseguimos separar
gdf_grande_sp.plot()
```

```python
# consertar: 
# "São Lourenço da Serra Suzano" -> "São Lourenço da Serra"
# "Santana do Parnaíba" -> Santana de Parnaíba"
# "Embu" -> "Embu das Artes"

lista_grande_sp = ["São Paulo","Arujá","Barueri","Biritiba Mirim","Caieiras","Cajamar",
                   "Carapicuíba","Cotia","Diadema","Embu das Artes","Embu-Guaçu","Ferraz de Vasconcelos",
                   "Francisco Morato","Franco da Rocha","Guararema","Guarulhos","Itapecerica da Serra",
                   "Itapevi","Itaquaquecetuba","Jandira","Juquitiba","Mairiporã","Mauá","Mogi das Cruzes",
                   "Osasco","Pirapora do Bom Jesus","Poá","Ribeirão Pires","Rio Grande da Serra","Salesópolis",
                   "Santa Isabel","Santana de Parnaíba","Santo André","São Bernardo do Campo",
                   "São Caetano do Sul","São Lourenço da Serra","Suzano","Taboão da Serra",
                   "Vargem Grande Paulista"]
```

```python
# salvando os dados de geometria da grande sp
gdf_grande_sp.to_file("dados/municipios_separados_grande_sp.json", driver="GeoJSON")
```

### Unindo todos os municípios eu uma geometria

```python
gdf_grande_sp = gdf_grande_sp.dissolve(by='SIGLA_UF')
gdf_grande_sp.plot()
```

```python
# vamos primeiro pegar o polygon de são paulo
polygon_grande_sp = gdf_grande_sp.iloc[0].geometry

# vamos separar os dados de roubo registrados em grande são paulo
gdf_roubos_grande_sp = gdf_roubos[gdf_roubos.intersects(polygon_grande_sp)]
```
